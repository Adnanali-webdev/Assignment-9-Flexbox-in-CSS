Q1)  Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When would you choose to use one over the other?

Ans) The basic difference between CSS grid layout and CSS flexbox layout is that flexbox was designed for layout in one dimension - either a row or a column. Grid was designed for two-dimensional layout - rows, and columns at the same time.

  

  | Feature |**Flexbox**|**Grid**|
  | --------|-----------|--------|
| **Layout Axis** | One-dimensional (either **row** or **column**)| Two-dimensional (**rows and columns**)|
| **Primary Use Case**   | Distributing space and aligning items along a single axis      | Designing complex grid-based layouts                              |
| **Item Placement**     | Items are placed in order and flow naturally in one direction     | Items can be placed precisely anywhere within the grid            |
| **Control**            | More control over **individual items**                            | More control over the **overall layout structure**                |
| **Responsiveness**     | Great for small-scale layouts like navigation bars or form fields | Ideal for large-scale layouts like entire web pages or dashboards |
| **Implicit Structure** | Structure defined by content flow                                 | Explicit structure with defined rows and columns                  |
| **Alignment Options**  | Powerful alignment tools along main and cross axes                | Offers full alignment control on both axes with finer granularity |


Q2) Explain the role of the following key properties in the Flexbox layout model

1 justify-content
2 align-items
3 gap
4 flex-direction
5 flex-wrap

Ans)
1 Justify-content : This property help us to align and distribute elements across the main axis.

common value (flex-start,flex-end,center,space-between,space-evenly,space-around)

2 align-items : Controls how flex items are aligned along the cross axis (perpendicular to the main axis).

common value(flex-start,flex-end,center,baseline,stretch)

3 gap : Specifies space between flex items.

4 flex-direction : This property specifies the direction of the  flex-item inside the flex-container.

or

Defines the direction of the main axis, which determines how items are laid out.

value ( column, row, column-reverse, row-reverse ).

5 flex-wrap : the property allow flex-items to wrap into multiple line.

or

Controls whether flex items are allowed to wrap onto multiple lines when there's not enough space.

value(wrap, nowrap, wrap-reverse)
